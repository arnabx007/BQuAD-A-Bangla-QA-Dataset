```python
import json

bquad_path = r'./bquad.json'

with open(bquad_path, 'r', encoding='UTF-8') as f:
    bquad = json.load(f)
```

```python
bquad.keys()
```

## Bquad contains around 13k QA pairs from 6000+ articles of 1000+ titles.

```python
data = bquad['data']

print(f'Titles: {len(data)}')

paras, qas = 0, 0

for i in range(len(data)):
    paras+= len(data[i]['paragraphs'])
    for j in range(len(data[i]['paragraphs'])):
        qas+= len(data[i]['paragraphs'][j]['qas'])

print(f'Paragraphs: {paras}')
print(f'QA Pairs: {qas}')
```

```python
titles = [data[i]['title'] for i in range(len(data))]
print(titles)
```
